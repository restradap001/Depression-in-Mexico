This query searches for any tweet posted in Mexico City since June 1, 2021 within a 10-kilometer radius.

```python
c = twint.Config()
c.Geo = "19.4978, -99.1269, 10km"
c.Since = "2021-06-01"
c.Store_csv = True
c.Output = "C:/Users/Lordd/Desktop/Depression/twitter-data-raw.csv"
c.Hide_output = True
```
